```python
import PyPDF2
import re

# Function to extract sentences from text
def extract_sentences(text):
    sentences = re.split(r'(?<!\w\.\w.)(?<![A-Z][a-z]\.)(?<=\.|\?|\!)\s', text)
    return sentences

# Function to extract text from a PDF
def extract_text_from_pdf(pdf_file):
    with open(pdf_file, 'rb') as file:
        pdf_reader = PyPDF2.PdfReader(file)
        pdf_text = ""
        for page_num in range(len(pdf_reader.pages)):
            page = pdf_reader.pages[page_num]
            pdf_text += page.extract_text()
    return pdf_text

# PDF file input and output text file
pdf_file = "F:\Download Folder\Deciphering RNA splicing logic with interpretable machine learning.pdf"
output_text_file = 'output_sentences.txt'

# Extract text from the PDF
pdf_text = extract_text_from_pdf(pdf_file)

# Extract sentences from the text
sentences = extract_sentences(pdf_text)

# Save the extracted sentences to a text file
with open(output_text_file, 'w', encoding='utf-8') as output_file:
    for sentence in sentences:
        output_file.write(sentence + '\n')

print(f'Sentences extracted and saved to {output_text_file}')
```
