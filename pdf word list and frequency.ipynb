```python
import os
import PyPDF2
import csv
import string
import re
from collections import Counter

# Load PDF file
pdfFileObj = open('F:\Download Folder\Jerusalem_Karen Armstrong.pdf', 'rb')
pdfReader = PyPDF2.PdfReader(pdfFileObj)

# Extract text from all pages
pdfText = ""
for pageNum in range(len(pdfReader.pages)):
    pageObj = pdfReader.pages[pageNum]
    pdfText += pageObj.extract_text()

# Convert to lowercase    
pdfText = pdfText.lower() 

# Remove punctuation marks
pdfText = pdfText.translate(str.maketrans("", "", string.punctuation))

# Remove non-English symbols
pdfText = re.sub(r'[^a-zA-Z\s]', '', pdfText)

# Split into words
words = pdfText.split()  

# Read excluded words from CSV file
excluded_words = []
with open('F://Download Folder//excludedwords.csv', 'r', encoding='utf-8') as csvFile:
    csvReader = csv.reader(csvFile)
    for row in csvReader:
        excluded_words.extend(row)

# Remove excluded words and single-letter words
filtered_words = [word for word in words if word not in excluded_words and len(word) > 2]

# Count word frequencies
wordCount = Counter(filtered_words)

# Define the output directory and file path
output_directory = 'F://Download Folder'
output_csv_path = os.path.join(output_directory, 'word_countbook.csv')

# Create the output directory if it doesn't exist
if not os.path.exists(output_directory):
    os.makedirs(output_directory)

# Create CSV file and write rows
with open(output_csv_path, 'w', newline='', encoding='utf-8') as csvFile:
    csvWriter = csv.writer(csvFile)

    for word, count in wordCount.most_common():
        row = [word, count]
        csvWriter.writerow(row)


### Features: Length more than 2, no non_english symbol, a list is needed to exclude which words
```
