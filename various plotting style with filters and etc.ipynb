```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Replace 'data.csv' with the actual path to your CSV file
data = pd.read_csv('F://Download Folder//fig16.csv')
# Assuming your X column is named 'x' and Y column is named 'y'
x = data['x']
y = data['y']

# Interpolate the data using numpy's interp function
x_interp = np.linspace(x.min(), x.max(), num=len(x)*10)  # Increase the number of points
y_interp = np.interp(x_interp, x, y)

plt.figure(figsize=(10, 6))  # Optional: Set the figure size
plt.plot(x_interp, y_interp, marker='none', linestyle='-', linewidth=1.2)

plt.grid(True, color='0.7', linestyle='-', linewidth=0.5)  # Use a darker gray color

plt.title('X vs Y with Linear Interpolation')
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.show()
```

```python
### Moving average smoothing
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Replace 'data.csv' with the actual path to your CSV file
data = pd.read_csv('F://Download Folder//fig16.csv')
# Assuming your X column is named 'x' and Y column is named 'y'
x = data['x']
y = data['y']

# Apply a moving average filter to smooth out the data
window_size = 5  # Adjust the window size as needed
y_smoothed = np.convolve(y, np.ones(window_size)/window_size, mode='same')

plt.figure(figsize=(10, 6))  # Optional: Set the figure size
plt.plot(x, y_smoothed, marker='none', linestyle='-', linewidth=1.2)

plt.grid(True, color='0.7', linestyle='-', linewidth=0.5)  # Use a darker gray color

plt.title('X vs Y with Moving Average Smoothing')
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline

# Replace 'data.csv' with the actual path to your CSV file
data = pd.read_csv('F://Download Folder//fig16.csv')
# Assuming your X column is named 'x' and Y column is named 'y'
x = data['x']
y = data['y']

# Use cubic spline interpolation to smooth out the data
cs = CubicSpline(x, y)

# Generate denser x values for smoother curve
x_dense = np.linspace(x.min(), x.max(), num=1000)

plt.figure(figsize=(10, 6))  # Optional: Set the figure size
plt.plot(x_dense, cs(x_dense), linestyle='-', linewidth=1.2)

plt.grid(True, color='0.7', linestyle='-', linewidth=0.5)  # Use a darker gray color

plt.title('X vs Y with Cubic Spline Interpolation')
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.show()
```

```python
##### Monotone Cubic Spline Interpolation
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import PchipInterpolator

# Replace 'data.csv' with the actual path to your CSV file
data = pd.read_csv('F://Download Folder//fig16.csv')
# Assuming your X column is named 'x' and Y column is named 'y'
x = data['x']
y = data['y']

# Create a PchipInterpolator object
interp_func = PchipInterpolator(x, y)

# Generate denser x values for plotting
x_dense = np.linspace(x.min(), x.max(), num=1000)

# Interpolate y values using the PchipInterpolator
y_interp = interp_func(x_dense)

plt.figure(figsize=(10, 6))  # Optional: Set the figure size
plt.plot(x_dense, y_interp, linestyle='-', linewidth=1.2)

plt.grid(True, color='0.7', linestyle='-', linewidth=0.5)  # Use a darker gray color

plt.title('X vs Y with Monotone Cubic Spline Interpolation (PchipInterpolator)')
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.show()

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import savgol_filter

# Replace 'data.csv' with the actual path to your CSV file
data = pd.read_csv('F://Download Folder//fig16.csv')
# Assuming your X column is named 'x' and Y column is named 'y'
x = data['x']
y = data['y']

# Apply Savitzky-Golay filter for smoothing
window_length = 15  # Adjust the window length as needed
y_smooth = savgol_filter(y, window_length, 3)  # 3rd order polynomial

plt.figure(figsize=(10, 6))  # Optional: Set the figure size
plt.plot(x, y_smooth, linestyle='-', linewidth=1.2)

plt.grid(True, color='0.7', linestyle='-', linewidth=0.5)  # Use a darker gray color

plt.title('X vs Smoothed Y using Savitzky-Golay Filter')
plt.xlabel('X axis')
plt.ylabel('Smoothed Y axis')

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Replace 'data.csv' with the actual path to your CSV file
data = pd.read_csv('F://Download Folder//fig16.csv')
# Assuming your X column is named 'x' and Y column is named 'y'
x = data['x']
y = data['y']

# Apply Exponential Moving Average (EMA) for smoothing
span = 500  # Adjust the span parameter as needed
y_smooth = pd.Series(y).ewm(span=span, adjust=False).mean()

plt.figure(figsize=(10, 6))  # Optional: Set the figure size
plt.plot(x, y_smooth, linestyle='-', linewidth=1.2)

plt.grid(True, color='0.7', linestyle='-', linewidth=0.5)  # Use a darker gray color

plt.title('X vs Smoothed Y using Exponential Moving Average (EMA) Filter')
plt.xlabel('X axis')
plt.ylabel('Smoothed Y axis')

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pykalman import KalmanFilter

# Replace 'data.csv' with the actual path to your CSV file
data = pd.read_csv('F://Download Folder//fig16.csv')
# Assuming your X column is named 'x' and Y column is named 'y'
x = data['x']
y = data['y']

# Apply Kalman filtering
kf = KalmanFilter(initial_state_mean=y[0], n_dim_obs=1)
y_smoothed, _ = kf.smooth(y)

plt.figure(figsize=(10, 6))  # Optional: Set the figure size
plt.plot(x, y_smoothed, linestyle='-', linewidth=1.2)

plt.grid(True, color='0.7', linestyle='-', linewidth=0.5)  # Use a darker gray color

plt.title('X vs Smoothed Y using Kalman Filter')
plt.xlabel('X axis')
plt.ylabel('Smoothed Y axis')

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, lfilter

# Replace 'data.csv' with the actual path to your CSV file
data = pd.read_csv('F://Download Folder//fig16.csv')
# Assuming your X column is named 'x' and Y column is named 'y'
x = data['x']
y = data['y']

# Apply Butterworth low-pass filter for smoothing
cutoff_freq = 0.1  # Adjust the cutoff frequency as needed
b, a = butter(N=4, Wn=cutoff_freq, btype='low')
y_smooth = lfilter(b, a, y)

plt.figure(figsize=(10, 6))  # Optional: Set the figure size
plt.plot(x, y_smooth, linestyle='-', linewidth=1.2)

plt.grid(True, color='0.7', linestyle='-', linewidth=0.5)  # Use a darker gray color

plt.title('X vs Smoothed Y using Butterworth Filter')
plt.xlabel('X axis')
plt.ylabel('Smoothed Y axis')

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from skimage.restoration import denoise_tv_chambolle

# Replace 'data.csv' with the actual path to your CSV file
data = pd.read_csv('F://Download Folder//fig16.csv')
# Assuming your X column is named 'x' and Y column is named 'y'
x = data['x']
y_noisy = data['y'] + np.random.normal(0, 0.2, len(data))  # Adding random noise

# Apply Total Variation Denoising for noise reduction
y_denoised = denoise_tv_chambolle(y_noisy, weight=0.1)

plt.figure(figsize=(10, 6))  # Optional: Set the figure size
# plt.plot(x, y_noisy, 'b', alpha=0.5, label='Noisy Data')
plt.plot(x, y_denoised, 'r', linestyle='-', linewidth=1.2, label='Denoised Data')

plt.grid(True, color='0.7', linestyle='-', linewidth=0.5)  # Use a darker gray color
plt.legend()

plt.title('Noisy Data vs Denoised Data using Total Variation Denoising')
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Replace 'data.csv' with the actual path to your CSV file
data = pd.read_csv('F://Download Folder//figsixteen.csv')
# Assuming your X column is named 'x' and Y column is named 'y'
x = data['X']
y = data['Y']
lowess = sm.nonparametric.lowess(y, x, frac=0.1)  # Adjust the 'frac' parameter as needed
plt.figure(figsize=(10, 6))  # Optional: Set the figure size
plt.plot(lowess[:, 0], lowess[:, 1], 'r', linestyle='-', linewidth=1.2, label='Lowess Smoothed')
plt.grid(True, color='0.8', linestyle='-', linewidth=0.5)  # Use a darker gray color
plt.show()
```
