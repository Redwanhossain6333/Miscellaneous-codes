```python
import matplotlib.pyplot as plt
import pandas as pd
from pywaffle import Waffle

# Load CSV file into a Pandas DataFrame
csv_file_path = "F:\Download Folder\export data.csv"  # Replace with the actual path to your CSV file
df = pd.read_csv(csv_file_path)

# Extract data from DataFrame
categories = df['Destination Country'].tolist()
values = df['Export (in USD)'].tolist()

# Create a dictionary to store category-value pairs
data = dict(zip(categories, values))

# Create a waffle chart
fig = plt.figure(
    FigureClass=Waffle,
    rows=150,  # Number of rows in the chart
    columns=100,  # Number of columns in the chart
    values=data,  # Category-value pairs
    legend={
        'loc': 'best',
        'bbox_to_anchor': (1, 1),
        'ncol': 1,  # Number of columns in the legend
        'fontsize': 8.2  # Font size of the legend
    },
    figsize=(35, 10)  # Size of the chart
)

# Add title to the chart
plt.title('Comparative Analysis of Export of Bangladesh to various Countries')

# Show the chart
plt.show()
```
